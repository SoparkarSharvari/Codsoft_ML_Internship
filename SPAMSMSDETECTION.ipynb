```python
data=data.drop(['Unnamed: 2','Unnamed: 3','Unnamed: 4'],axis='columns')
```

```python
#rename v1 column to target and v2 column to message
data.rename(columns=dict({"v1":"target","v2":"message"}),inplace=True)
```

```python
#after deleting and renaming columns 
# Assuming 'text' is the column containing SMS messages and 'label' is the column with spam or ham labels
spam_messages = data[data['target'] == 'spam']['message']

# Print the first few spam messages
for idx, spam_message in enumerate(spam_messages.head()):
    print(f"Spam Message {idx + 1}:\n{spam_message}\n{'-'*50}")
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(data['message'], data['target'], test_size=0.2, random_state=42)

# Create TF-IDF vectorizer
tfidf_vectorizer = TfidfVectorizer(stop_words='english', max_features=5000)
X_train_tfidf = tfidf_vectorizer.fit_transform(X_train)
X_test_tfidf = tfidf_vectorizer.transform(X_test)

# Train and evaluate Naive Bayes classifier
nb_classifier = MultinomialNB()
nb_classifier.fit(X_train_tfidf, y_train)
nb_predictions = nb_classifier.predict(X_test_tfidf)

print("Naive Bayes Classifier:")
print("Accuracy:", accuracy_score(y_test, nb_predictions))
print("Confusion Matrix:\n", confusion_matrix(y_test, nb_predictions))
print("Classification Report:\n", classification_report(y_test, nb_predictions))
```
